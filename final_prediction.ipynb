```
import os
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
```

```
my_Model = load_model('/content/drive/My Drive/model.h5')
```

```
def predict_(filename):
    test_image = image.load_img(filename,
                                target_size = (128, 128))
    test_image = image.img_to_array(test_image)
    
    test_image = np.expand_dims(test_image, axis = 0)

    result = my_Model.predict(test_image)
    
    test_image = np.squeeze(test_image)
    
    plt.title("{}".format("Dog" if result[0][1] == 1 else "Cat"))
    print(type(result))
    plt.imshow(test_image/255.)
```

```
filename = random.choice(os.listdir('/content/drive/My Drive/test'))
predict_("/content/drive/My Drive/test/{}".format(filename))
```
